## Problem 1 — Inspect Built-in Objects

For each value below, print:

- the value
- its type
- whether its type is itself an object
- whether the value has a `__class__` attribute

````python
values = [10, 10.5, True, None, "python", [1, 2, 3], {"a": 1}, int, list]

for value in values:
    print("value:", repr(value))
    print("type:", type(value))
    print("type is object:", isinstance(type(value), object))
    print("has __class__:", hasattr(value, "__class__"))
    print("-" * 40)
````

### Solution 1

Every value in the list is an object.

Even classes such as `int` and `list` are objects. Their type is usually `type`.

A key idea is that objects have types, and types are objects too.

## Problem 3 — Function Assignment and Identity

Predict the output before running the cell.

Does assigning a function to another variable copy the function?

````python
def double(x):
    return x * 2

f = double

print(f is double)
print(f(10))
print(double(10))
print(type(f))
````

### Solution 3

`f = double` does not copy the function.

Therefore:

```python
f is double
```

is `True`.

## Problem 4 — Return a Function

Complete `select_operation` so it returns one of three functions:

- `square`
- `cube`
- `negate`

Then call the returned function.

````python
def square(x):
    return x ** 2

def cube(x):
    return x ** 3

def negate(x):
    return -x

def select_operation(name):
    if name == "square":
        return square
    if name == "cube":
        return cube
    if name == "negate":
        return negate
    raise ValueError(f"Unknown operation: {name}")

operation = select_operation("cube")

print(operation)
print(operation is cube)
print(operation(3))
print(select_operation("square")(5))
````

### Solution 4

A function can return another function because functions are objects.

`select_operation("cube")` returns the function object named `cube`.

That returned object can be stored in a variable or called immediately.

## Problem 5 — Pass a Function as an Argument

Write a function named `apply_twice` that receives a function and a value.

It should apply the function twice.

Example:

```python
apply_twice(double, 3)
```

should return `12`, because `double(double(3)) == 12`.

````python
def double(x):
    return x * 2

def increment(x):
    return x + 1

def apply_twice(fn, value):
    return fn(fn(value))

print(apply_twice(double, 3))
print(apply_twice(increment, 3))
````

### Solution 5

`fn` is a variable that refers to a function object.

Because function objects are callable, `fn(value)` works just like calling the original function by name.

## Problem 6 — Build a Function Registry

Create a dictionary that maps operation names to function objects.

Then write `calculate(name, value)` that looks up the function and calls it.

````python
def square(x):
    return x ** 2

def cube(x):
    return x ** 3

def half(x):
    return x / 2

operations = {
    "square": square,
    "cube": cube,
    "half": half,
}

def calculate(name, value):
    if name not in operations:
        raise ValueError(f"Unsupported operation: {name}")
    return operations[name](value)

print(calculate("square", 6))
print(calculate("cube", 3))
print(calculate("half", 10))
````

### Solution 6

Dictionaries can store functions as values because functions are objects.

This avoids long chains of `if` / `elif` statements and is a common pattern for dispatching behavior.

## Problem 7 — Function Metadata

Functions carry metadata.

Inspect the following attributes:

- `__name__`
- `__doc__`
- `__annotations__`

````python
def area_rectangle(width: float, height: float) -> float:
    """Return the area of a rectangle."""
    return width * height

print(area_rectangle.__name__)
print(area_rectangle.__doc__)
print(area_rectangle.__annotations__)
print(area_rectangle(4, 5))
````

### Solution 7

A function is not just executable code.

It is an object with attributes.

This is why tools can inspect functions, generate documentation, read annotations, and decorate behavior.

## Problem 9 — Callable Objects

Create a class called `Multiplier` whose instances can be called like functions.

Example:

```python
triple = Multiplier(3)
triple(10)
```

should return `30`.

````python
class Multiplier:
    def __init__(self, factor):
        self.factor = factor

    def __call__(self, value):
        return self.factor * value

triple = Multiplier(3)
double = Multiplier(2)

print(triple(10))
print(double(10))
print(callable(triple))
print(type(triple))
````

### Solution 9

An object is callable when its class defines `__call__`.

Functions are callable objects, but they are not the only callable objects.

## Problem 10 — Higher-Order Function Factory

Write a function named `make_power` that returns a new function.

The returned function should raise its argument to the selected exponent.

````python
def make_power(exponent):
    def power(value):
        return value ** exponent
    return power

square = make_power(2)
cube = make_power(3)

print(square(5))
print(cube(5))
print(square is cube)
print(type(square))
````

### Solution 10

`make_power` creates and returns a new function object.

Each returned function remembers the `exponent` value from the call that created it.

This is called a closure.

## Problem 11 — Sort Using a Function Object

Sort the list of students by grade using a function passed as the `key` argument.

````python
students = [
    {"name": "Ada", "grade": 95},
    {"name": "Linus", "grade": 88},
    {"name": "Grace", "grade": 99},
]

def get_grade(student):
    return student["grade"]

sorted_students = sorted(students, key=get_grade, reverse=True)

print(sorted_students)
````

### Solution 11

`sorted` receives a function object through `key`.

For each item, it calls that function to decide which value should be used for sorting.

## Problem 12 — Mini Pipeline Engine

Create a function named `pipeline` that receives:

- an initial value
- a list of functions

It should apply each function in order and return the final result.

````python
def strip_text(text):
    return text.strip()

def lowercase(text):
    return text.lower()

def replace_spaces(text):
    return text.replace(" ", "-")

def pipeline(value, functions):
    result = value
    for fn in functions:
        result = fn(result)
    return result

slug = pipeline("  Everything Is An Object  ", [strip_text, lowercase, replace_spaces])

print(slug)
````

### Solution 12

The list stores function objects.

The loop calls each function in sequence.

This pattern is common in data processing, validation, parsing, and transformation pipelines.
