## Best-practice reminders

- Prefer `elif` over deeply nested `if/else` chains when testing multiple exclusive cases.
- Use the ternary operator only for simple, readable expressions.
- Keep side effects out of conditions when possible.
- Write conditions that are easy to read before trying to make them shorter.
- Order conditions carefully: Python stops at the first matching branch.


## Problem 1 — Predict the Output of an `if / elif / else` Chain

Without running the code, determine what will be printed and explain why only one branch runs.

```python
score = 82

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("D or below")
```

### Solution 1

```python
score = 82

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
else:
    print("D or below")
```

**Explanation:**

`score >= 90` is `False`, but `score >= 80` is `True`, so Python executes that branch and stops checking the rest of the chain.

In an `if / elif / else` chain, only the **first matching branch** runs.

## Problem 2 — Fix Incorrect Branch Ordering

The goal is to classify a number as:

- `'negative'` if it is below 0
- `'small'` if it is from 0 to 9
- `'large'` if it is 10 or greater

The following code is incorrect. Fix it.

```python
x = 12

if x >= 0:
    print("small")
elif x >= 10:
    print("large")
else:
    print("negative")
```

### Solution 2

```python
x = 12

if x < 0:
    print("negative")
elif x < 10:
    print("small")
else:
    print("large")
```

**Explanation:**

The original code checked `x >= 0` first, which captures all non-negative values, including large ones.

Condition order matters. More specific logic must not be hidden behind broader conditions.

## Problem 3 — Rewrite Nested Conditionals Using `elif`

Rewrite the following code into a cleaner `if / elif / else` chain without changing behavior.

```python
age = 16

if age < 13:
    category = "child"
else:
    if age < 18:
        category = "teen"
    else:
        category = "adult"

print(category)
```

### Solution 3

```python
age = 16

if age < 13:
    category = "child"
elif age < 18:
    category = "teen"
else:
    category = "adult"

print(category)
```

**Why this is better:**

`elif` makes mutually exclusive branches easier to read than nested `if` blocks.

This is usually the preferred style for multi-branch classification.

## Problem 4 — Multi-Condition Decision

Write code that prints `'access granted'` only if all of the following are true:

- `age >= 18`
- `has_ticket` is `True`
- `banned` is `False`

Otherwise print `'access denied'`.

Use clear, best-practice formatting.

### Solution 4

```python
age = 21
has_ticket = True
banned = False

if age >= 18 and has_ticket and not banned:
    print("access granted")
else:
    print("access denied")
```

**Explanation:**

All three conditions must be true at the same time, so `and` is appropriate.

Using `not banned` is clearer than writing `banned == False`.

## Problem 5 — Ternary Operator for a Simple Value

Use a conditional expression to assign `'even'` to `label` when `n` is even, and `'odd'` otherwise.

### Solution 5

```python
n = 17
label = "even" if n % 2 == 0 else "odd"
print(label)
```

**Best-practice note:**

This is a good use of the ternary operator because the logic is short and easy to read.

## Problem 6 — When *Not* to Use the Ternary Operator

The code below works, but it is not very readable. Rewrite it using a regular `if / else` block.

```python
temperature = 31
message = "hot" if temperature > 30 else "warm" if temperature > 20 else "cool" if temperature > 10 else "cold"
print(message)
```

### Solution 6

```python
temperature = 31

if temperature > 30:
    message = "hot"
elif temperature > 20:
    message = "warm"
elif temperature > 10:
    message = "cool"
else:
    message = "cold"

print(message)
```

**Explanation:**

Nested ternary expressions are legal, but they often hurt readability.

For multiple branches, `if / elif / else` is usually the better choice.

## Problem 7 — Understand Evaluation Order

Predict what will be printed.

```python
value = 7

if value > 0:
    print("positive")

if value > 5:
    print("greater than 5")
else:
    print("5 or less")
```

### Solution 7

```python
value = 7

if value > 0:
    print("positive")

if value > 5:
    print("greater than 5")
else:
    print("5 or less")
```

**Explanation:**

These are two separate `if` statements, not one combined chain. The `else` belongs only to the second `if`.

So both `'positive'` and `'greater than 5'` are printed.

## Problem 8 — Short-Circuit Safety

The code below may raise an error. Rewrite the condition safely so it only checks the first character when the string is non-empty.

```python
name = ""

if name[0] == "A" and len(name) > 0:
    print("Starts with A")
else:
    print("Does not start with A")
```

### Solution 8

```python
name = ""

if len(name) > 0 and name[0] == "A":
    print("Starts with A")
else:
    print("Does not start with A")
```

**Explanation:**

Python evaluates `and` from left to right and stops as soon as the result is known.

By checking `len(name) > 0` first, the expression safely avoids `name[0]` when the string is empty.

## Problem 9 — Choose Between Two Functions

Use a conditional expression to call the correct function based on whether `is_morning` is `True`.

```python
def greet_morning():
    print("Good morning!")

def greet_evening():
    print("Good evening!")

is_morning = False
```

### Solution 9

```python
def greet_morning():
    print("Good morning!")

def greet_evening():
    print("Good evening!")

is_morning = False
greet_morning() if is_morning else greet_evening()
```

**Note:**

This is valid, but in production code many developers still prefer a plain `if / else` statement for function calls because it can be easier to scan.

## Problem 10 — Classify Using Multiple Ranges

Given `hour` in 24-hour time, assign:

- `'morning'` for `5 <= hour < 12`
- `'afternoon'` for `12 <= hour < 17`
- `'evening'` for `17 <= hour < 21`
- `'night'` otherwise

Then print the result.

### Solution 10

```python
hour = 19

if 5 <= hour < 12:
    period = "morning"
elif 12 <= hour < 17:
    period = "afternoon"
elif 17 <= hour < 21:
    period = "evening"
else:
    period = "night"

print(period)
```

**Why this is good Python:**

Chained comparisons like `5 <= hour < 12` are both readable and idiomatic.

## Problem 12 — Nested Conditional with a Meaningful Use Case

A store has this rule:

- if the customer is not a member, discount is `0`
- if the customer is a member and has spent at least `1000`, discount is `20`
- otherwise discount is `10`

Write a solution first using nested `if` statements.

### Solution 12

```python
is_member = True
total_spent = 1200

if is_member:
    if total_spent >= 1000:
        discount = 20
    else:
        discount = 10
else:
    discount = 0

print(discount)
```

Now rewrite the same logic using `if / elif / else` for flatter structure.

```python
is_member = True
total_spent = 1200

if not is_member:
    discount = 0
elif total_spent >= 1000:
    discount = 20
else:
    discount = 10

print(discount)
```

**Lesson:**

Nested conditionals are sometimes correct, but flatter logic is often easier to maintain.

## Challenge Problem 13 — Predict Function Evaluation in a Ternary Expression

Without running the code, determine which function is called and what is printed.

```python
def left():
    print("left called")
    return "LEFT"

def right():
    print("right called")
    return "RIGHT"

flag = True
result = left() if flag else right()
print(result)
```

### Solution 13

```python
def left():
    print("left called")
    return "LEFT"

def right():
    print("right called")
    return "RIGHT"

flag = True
result = left() if flag else right()
print(result)
```

**Explanation:**

Only the selected side of the conditional expression is evaluated. Since `flag` is `True`, `left()` runs and `right()` is never called.

## Challenge Problem 14 — Combine Conditionals with Return Values

Write a function `describe_number(n)` that returns:

- `'negative even'`
- `'negative odd'`
- `'zero'`
- `'positive even'`
- `'positive odd'`

depending on the value of `n`.

Use clean, readable conditional logic and then test it on `-3`, `0`, and `8`.

### Solution 14

```python
def describe_number(n):
    if n == 0:
        return "zero"
    elif n < 0:
        if n % 2 == 0:
            return "negative even"
        else:
            return "negative odd"
    else:
        if n % 2 == 0:
            return "positive even"
        else:
            return "positive odd"

print(describe_number(-3))
print(describe_number(0))
print(describe_number(8))
```

### Alternative flatter solution

```python
def describe_number(n):
    if n == 0:
        return "zero"

    parity = "even" if n % 2 == 0 else "odd"

    if n < 0:
        return f"negative {parity}"
    else:
        return f"positive {parity}"

print(describe_number(-3))
print(describe_number(0))
print(describe_number(8))
```
