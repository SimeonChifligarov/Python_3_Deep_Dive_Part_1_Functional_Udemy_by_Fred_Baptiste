## Problem 1 - Iteration versus repetition

These two code snippets both print numbers, but they are not conceptually the same. Explain the difference and predict the output.

````python
i = 0
while i < 4:
    print(i)
    i += 1

print('---')

for i in range(4):
    print(i)
````

### Solution 1

Expected output:

```text
0
1
2
3
---
0
1
2
3
```

### Explanation

Although both snippets produce the same output, the ideas are different:

- The `while` loop is **repetition based on a condition**.
- The `for` loop is **iteration over an iterable**.

In `for i in range(4)`, `range(4)` provides values one at a time. The loop does not manually manage `i`.

## Problem 2 - Predict iteration over different iterables

What is printed by this code?

````python
items = [
    [1, 2],
    'ab',
    (3, 4)
]

for item in items:
    print(item)
    for sub_item in item:
        print('  ', sub_item)
````

### Solution 2

Expected output:

```text
[1, 2]
   1
   2
ab
   a
   b
(3, 4)
   3
   4
```

### Explanation

Each outer element is itself iterable:

- `[1, 2]` yields `1`, then `2`
- `'ab'` yields `'a'`, then `'b'`
- `(3, 4)` yields `3`, then `4`

This highlights that Python's `for` loop works with many iterable types, not just lists.

## Problem 3 - Tuple unpacking in a `for` loop

Predict the output. Then explain why the loop variable on the left has two names.

````python
pairs = [(1, 10), (2, 20), (3, 30)]

for a, b in pairs:
    print(a + b)
````

### Solution 3

Expected output:

```text
11
22
33
```

### Explanation

On each iteration, the loop receives one tuple from `pairs`.

- `(1, 10)` becomes `a = 1`, `b = 10`
- `(2, 20)` becomes `a = 2`, `b = 20`
- `(3, 30)` becomes `a = 3`, `b = 30`

This is called **unpacking**. It is often cleaner than writing `for pair in pairs:` and then indexing with `pair[0]` and `pair[1]`.

## Problem 4 - When unpacking fails

What happens when you run this code?

````python
data = [(1, 2), (3, 4, 5), (6, 7)]

for x, y in data:
    print(x, y)
````

### Solution 4

The first iteration works, but the second one raises an exception.

Printed before the error:

```text
1 2
```

Then Python raises:

```text
ValueError: too many values to unpack (expected 2)
```

### Explanation

The loop expects every element of `data` to contain exactly two values. But `(3, 4, 5)` has three values, so unpacking fails.

### Best-practice note

Only unpack when the structure is guaranteed. Otherwise, validate or handle the shape explicitly.

## Problem 6 - `break` and loop `else`

What is the output? Explain why the `else` block does or does not run.

````python
for i in range(1, 8):
    print(i)
    if i == 5:
        print('breaking')
        break
else:
    print('completed normally')
````

### Solution 6

Expected output:

```text
1
2
3
4
5

```

### Explanation

The loop stops when `i == 5`, so the `else` block does **not** run.

### Rule

A loop `else` runs only if the loop finishes **without** hitting `break`.

## Problem 10 - Searching with `for ... else`

This is one of the most useful real-world uses of loop `else`. What is printed?

````python
target = 13
values = [2, 4, 6, 8, 10]

for value in values:
    if value == target:
        print('found')
        break
else:
    print('not found')
````

### Solution 10

Expected output:

```text
not found
```

### Explanation

The loop checks each value. Since `13` is never found, the loop never breaks, so the `else` clause runs.

This is a clean pattern for search logic.

## Problem 11 - Manual indexing versus `enumerate()`

All three approaches below aim to print indices and characters. Which is best, and why?

````python
s = 'python'

print('approach 1')
i = 0
for ch in s:
    print(i, ch)
    i += 1

print('approach 2')
for i in range(len(s)):
    print(i, s[i])

print('approach 3')
for i, ch in enumerate(s):
    print(i, ch)
````

### Solution 11

All three print the same pairs:

```text
0 p
1 y
2 t
3 h
4 o
5 n
```

### Best answer

**Approach 3** using `enumerate()` is generally best.

### Why

- Approach 1 works, but manual counter management is verbose and easier to get wrong.
- Approach 2 works for indexable sequences, but it is less direct and can be less readable.
- Approach 3 is explicit, concise, and Pythonic.

## Problem 12 - Refactor to a more Pythonic loop

Rewrite this code in a clearer, more idiomatic way.

````python
names = ['Ada', 'Grace', 'Linus']

for i in range(len(names)):
    print(i, names[i])
````

### Solution 12

A more Pythonic version is:

```python
names = ['Ada', 'Grace', 'Linus']

for i, name in enumerate(names):
    print(i, name)
```

### Why

If you need both index and value, `enumerate()` communicates that intent directly.

## Problem 13 - Nested `for` loops and `break`

Which loop does `break` exit here?

````python
for i in range(2):
    print(f'outer {i}')
    for j in range(5):
        if j == 2:
            break
        print(f'  inner {j}')
````

### Solution 13

Expected output:

```text
outer 0
  inner 0
  inner 1
outer 1
  inner 0
  inner 1
```

### Explanation

`break` exits only the **innermost** loop. So it stops the `for j in range(5)` loop, not the outer loop.

## Problem 15 - Trace a mixed-control-flow loop

What is the final value of `log`?

````python
log = []

for i in range(5):
    log.append(f'start-{i}')
    if i == 1:
        log.append('continue')
        continue
    if i == 3:
        log.append('break')
        break
    log.append(f'end-{i}')
else:
    log.append('loop-else')

log
````

### Solution 15

Final value of `log`:

```python
[
    'start-0', 'end-0',
    'start-1', 'continue',
    'start-2', 'end-2',
    'start-3', 'break'
]
```

### Explanation

- `i = 0`: normal iteration
- `i = 1`: `continue` skips the rest of the loop body
- `i = 2`: normal iteration
- `i = 3`: `break` exits the loop
- Because the loop ends with `break`, the loop `else` does not run

## Summary of best practices

1. Use `for` when iterating over values from an iterable.
2. Use `while` when repetition depends on a condition rather than an iterable.
3. Prefer unpacking when the iterable structure is known and stable.
4. Use `enumerate()` instead of manual counters when you need index and value.
5. Use loop `else` mainly for search-style logic where `break` means success or early exit.
6. Keep loop control flow readable; avoid cleverness that makes execution harder to trace.
