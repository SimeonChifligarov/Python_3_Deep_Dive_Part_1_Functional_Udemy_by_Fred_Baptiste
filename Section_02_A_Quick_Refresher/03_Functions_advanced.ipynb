## Problem 1 — Return values vs side effects

Write a function `analyze_numbers(nums)` that takes a list of numbers and returns a **dictionary** with:
- `count`
- `sum`
- `mean`
- `min`
- `max`

Requirements:
- The function must **return** the dictionary, not print it.
- If `nums` is empty, raise a `ValueError` with a helpful message.

```python
def analyze_numbers(nums):
    """Return summary statistics for a non-empty list of numbers."""
    if not nums:
        raise ValueError("nums must not be empty")
    
    total = sum(nums)
    return {
        "count": len(nums),
        "sum": total,
        "mean": total / len(nums),
        "min": min(nums),
        "max": max(nums),
    }


result = analyze_numbers([3, 7, 1, 9])
assert result == {
    "count": 4,
    "sum": 20,
    "mean": 5.0,
    "min": 1,
    "max": 9,
}
result
```

## Problem 2 — Flexible positional arguments with `*args`

Write a function `product_except_zero(*args)` that multiplies all non-zero numeric arguments together.

Requirements:
- Ignore any argument equal to `0`.
- If no non-zero values are provided, return `1`.
- Assume inputs are numbers.

```python
def product_except_zero(*args):
    """Multiply all non-zero arguments; return 1 if none qualify."""
    result = 1
    for value in args:
        if value != 0:
            result *= value
    return result


assert product_except_zero(2, 3, 0, 4) == 24
assert product_except_zero(0, 0, 0) == 1
assert product_except_zero() == 1
product_except_zero(5, 0, -2)
```

## Problem 3 — Keyword arguments with `**kwargs`

Write a function `build_profile(name, **kwargs)` that returns a dictionary representing a user profile.

Requirements:
- The returned dictionary must always include the key `name`.
- Any extra keyword arguments should also be included.
- If `name` is empty, raise a `ValueError`.

```python
def build_profile(name, **kwargs):
    """Return a profile dictionary with a required name field."""
    if not name:
        raise ValueError("name must not be empty")
    
    profile = {"name": name}
    profile.update(kwargs)
    return profile


profile = build_profile("Mina", age=21, city="Sofia", active=True)
assert profile == {
    "name": "Mina",
    "age": 21,
    "city": "Sofia",
    "active": True,
}
profile
```

## Problem 4 — Default arguments and good practice

Write a function `greet(name, greeting="Hello", punctuation="!")` that returns a greeting string.

Examples:
- `greet("Ana")` → `"Hello, Ana!"`
- `greet("Ana", greeting="Hi")` → `"Hi, Ana!"`
- `greet("Ana", punctuation=".")` → `"Hello, Ana."`

```python
def greet(name, greeting="Hello", punctuation="!"):
    """Return a formatted greeting."""
    return f"{greeting}, {name}{punctuation}"


assert greet("Ana") == "Hello, Ana!"
assert greet("Ana", greeting="Hi") == "Hi, Ana!"
assert greet("Ana", punctuation=".") == "Hello, Ana."
greet("Ivan", greeting="Welcome", punctuation="!!!")
```

## Problem 5 — Avoiding mutable default argument bugs

A common mistake is using a mutable object like `[]` as a default argument.

Write a function `append_item(item, target=None)` that appends `item` to a list and returns the list.

Requirements:
- If `target` is `None`, create a new empty list.
- Do **not** use `[]` directly as the default value.

```python
def append_item(item, target=None):
    """Append an item to a list without sharing state across calls."""
    if target is None:
        target = []
    target.append(item)
    return target


a = append_item(1)
b = append_item(2)
assert a == [1]
assert b == [2]
assert a is not b
append_item(3, [10, 20])
```

## Problem 6 — Functions as first-class objects

Write a function `apply_twice(func, value)` that applies a function to a value twice.

Example:
- if `func(x) = x + 1`, then `apply_twice(func, 3)` should return `5`.

```python
def apply_twice(func, value):
    """Apply func to value two times."""
    return func(func(value))


def add_one(x):
    return x + 1


assert apply_twice(add_one, 3) == 5
assert apply_twice(lambda x: x * 2, 5) == 20
apply_twice(abs, -7)
```

## Problem 7 — Returning functions (closures)

Write a function `make_power(exponent)` that returns a new function.

The returned function should take one argument `x` and return `x ** exponent`.

Example:
- `square = make_power(2)`
- `square(5)` → `25`

```python
def make_power(exponent):
    """Return a function that raises its input to the given exponent."""
    def power(x):
        return x ** exponent
    return power


square = make_power(2)
cube = make_power(3)

assert square(5) == 25
assert cube(2) == 8
square(9)
```

## Problem 9 — Input validation and annotations

Write a function

`safe_divide(a: float, b: float) -> float`

Requirements:
- Return `a / b`
- If `b == 0`, raise `ZeroDivisionError` with a custom message
- Include a docstring and type annotations

```python
def safe_divide(a: float, b: float) -> float:
    """Return a divided by b, raising a clear error when b is zero."""
    if b == 0:
        raise ZeroDivisionError("b must not be zero")
    return a / b


assert safe_divide(10, 2) == 5.0
assert safe_divide(7, 2) == 3.5
safe_divide(9, 3)
```

## Problem 11 — Recursion: factorial

Write a recursive function `factorial(n)`.

Requirements:
- `factorial(0)` should return `1`
- for negative inputs, raise `ValueError`
- use recursion, not a loop

```python
def factorial(n):
    """Return n! using recursion."""
    if n < 0:
        raise ValueError("n must be non-negative")
    if n == 0:
        return 1
    return n * factorial(n - 1)


assert factorial(0) == 1
assert factorial(1) == 1
assert factorial(5) == 120
factorial(6)
```

## Problem 12 — Recursion with memoization pattern

Write a function `fib(n, memo=None)` that returns the nth Fibonacci number.

Requirements:
- `fib(0) == 0`
- `fib(1) == 1`
- use recursion
- avoid repeated work by storing previously computed results in a dictionary

```python
def fib(n, memo=None):
    """Return the nth Fibonacci number using recursive memoization."""
    if n < 0:
        raise ValueError("n must be non-negative")
    if memo is None:
        memo = {}
    if n in memo:
        return memo[n]
    if n == 0:
        return 0
    if n == 1:
        return 1
    memo[n] = fib(n - 1, memo) + fib(n - 2, memo)
    return memo[n]


assert fib(0) == 0
assert fib(1) == 1
assert fib(10) == 55
fib(20)
```

## Problem 13 — Higher-order filtering

Write a function `filter_list(values, predicate)` that returns a new list containing only the items for which `predicate(item)` is `True`.

Do not use Python's built-in `filter`.

```python
def filter_list(values, predicate):
    """Return items from values that satisfy predicate."""
    result = []
    for item in values:
        if predicate(item):
            result.append(item)
    return result


nums = [1, 2, 3, 4, 5, 6]
evens = filter_list(nums, lambda x: x % 2 == 0)
big = filter_list(nums, lambda x: x > 3)

assert evens == [2, 4, 6]
assert big == [4, 5, 6]
evens
```

## Problem 14 — Composing functions

Write a function `compose(f, g)` that returns a new function equivalent to:

`h(x) = f(g(x))`

Then use it to compose:
- `double(x)` → `x * 2`
- `minus_three(x)` → `x - 3`

```python
def compose(f, g):
    """Return the composition f(g(x))."""
    def inner(x):
        return f(g(x))
    return inner


def double(x):
    return x * 2


def minus_three(x):
    return x - 3


combined = compose(double, minus_three)

assert combined(10) == 14
assert combined(3) == 0
combined(8)
```

## Problem 15 — Function decorators by hand

Write a decorator `trace_calls` that wraps a function so that each call returns a tuple:

`(result, message)`

where `message` is a string like:
`"Called add with args=(2, 3), kwargs={}"`

Apply it to a function `add(a, b)`.

```python
def trace_calls(func):
    """Wrap a function and return (result, trace_message) on each call."""
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        message = f"Called {func.__name__} with args={args}, kwargs={kwargs}"
        return result, message
    return wrapper


@trace_calls
def add(a, b):
    return a + b


value, msg = add(2, 3)
assert value == 5
assert msg == "Called add with args=(2, 3), kwargs={}"
(value, msg)
```

## Problem 16 — Practical mini-project

Write a function `make_report_card(student_name, scores)` where:
- `student_name` is a string
- `scores` is a dictionary such as `{"math": 91, "cs": 87, "history": 95}`

Return a dictionary with:
- `student`
- `subjects`
- `average`
- `best_subject`
- `worst_subject`
- `passed`

Rules:
- `passed` is `True` only if every score is at least 50
- if `scores` is empty, raise `ValueError`

```python
def make_report_card(student_name, scores):
    """Return a summary report card for a student."""
    if not scores:
        raise ValueError("scores must not be empty")
    
    average = sum(scores.values()) / len(scores)
    best_subject = max(scores, key=scores.get)
    worst_subject = min(scores, key=scores.get)
    passed = all(score >= 50 for score in scores.values())
    
    return {
        "student": student_name,
        "subjects": len(scores),
        "average": average,
        "best_subject": best_subject,
        "worst_subject": worst_subject,
        "passed": passed,
    }


card = make_report_card("Elena", {"math": 91, "cs": 87, "history": 95})
assert card["student"] == "Elena"
assert card["subjects"] == 3
assert round(card["average"], 2) == 91.0
assert card["best_subject"] == "history"
assert card["worst_subject"] == "cs"
assert card["passed"] is True
card
```
