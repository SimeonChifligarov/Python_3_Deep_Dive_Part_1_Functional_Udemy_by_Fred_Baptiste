## Problem 1 — Sum until threshold

Write a function `steps_to_reach(limit)` that starts from `total = 0` and repeatedly adds the next positive integer:

- add `1`, then `2`, then `3`, and so on
- stop once `total >= limit`
- return a tuple `(steps, total)`

Example:
- if `limit = 10`, the additions are `1 + 2 + 3 + 4 = 10`, so return `(4, 10)`

```python
def steps_to_reach(limit):
    """Return how many consecutive positive integers are needed to reach at least limit."""
    total = 0
    next_value = 1
    steps = 0
    
    while total < limit:
        total += next_value
        next_value += 1
        steps += 1
    
    return steps, total


assert steps_to_reach(1) == (1, 1)
assert steps_to_reach(10) == (4, 10)
assert steps_to_reach(11) == (5, 15)
steps_to_reach(25)
```

## Problem 2 — Count digits using a `while` loop

Write a function `digit_count(n)` that returns the number of digits in a non-negative integer using a `while` loop.

Requirements:
- `digit_count(0)` should return `1`
- do not convert the number to a string
- if `n < 0`, raise `ValueError`

```python
def digit_count(n):
    """Return the number of digits in a non-negative integer."""
    if n < 0:
        raise ValueError("n must be non-negative")
    if n == 0:
        return 1
    
    count = 0
    while n > 0:
        n //= 10
        count += 1
    return count


assert digit_count(0) == 1
assert digit_count(7) == 1
assert digit_count(12345) == 5
digit_count(1000000)
```

## Problem 3 — Reverse an integer without strings

Write a function `reverse_int(n)` that returns the digits of a non-negative integer in reverse order.

Examples:
- `reverse_int(1234)` → `4321`
- `reverse_int(1200)` → `21`

Requirements:
- use a `while` loop
- do not convert to a string
- raise `ValueError` for negative input

```python
def reverse_int(n):
    """Return a non-negative integer with its digits reversed."""
    if n < 0:
        raise ValueError("n must be non-negative")
    
    reversed_num = 0
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10
    return reversed_num


assert reverse_int(0) == 0
assert reverse_int(1234) == 4321
assert reverse_int(1200) == 21
reverse_int(90700)
```

## Problem 4 — Input validation pattern (simulated)

In real programs, `while` loops are often used to keep asking for input until valid data is received.

Here, instead of calling `input()`, write a function `first_valid_name(candidates)` that scans a list of attempted names and returns the first valid one.

A valid name must:
- have length at least 2
- contain only alphabetic characters
- be printable

If no valid name exists, return `None`.

```python
def first_valid_name(candidates):
    """Return the first valid name from a sequence of attempted inputs."""
    idx = 0
    while idx < len(candidates):
        name = candidates[idx]
        if len(name) >= 2 and name.isalpha() and name.isprintable():
            return name
        idx += 1
    return None


assert first_valid_name(["", "1", "A", "Mila"]) == "Mila"
assert first_valid_name(["!", "7x", "_"]) is None
first_valid_name(["Z", "Bo", "123"])
```

## Problem 5 — Sentinel-controlled loop

Write a function `collect_until_stop(items)` that processes values from a list until the sentinel value `'STOP'` appears.

Requirements:
- return a new list containing all items before `'STOP'`
- if `'STOP'` never appears, return all items
- use a `while` loop

```python
def collect_until_stop(items):
    """Collect items up to but not including the sentinel 'STOP'."""
    result = []
    idx = 0
    
    while idx < len(items):
        if items[idx] == "STOP":
            break
        result.append(items[idx])
        idx += 1
    
    return result


assert collect_until_stop([1, 2, 3, "STOP", 4]) == [1, 2, 3]
assert collect_until_stop(["a", "b"]) == ["a", "b"]
collect_until_stop([10, 20, "STOP", 30, 40])
```

## Problem 6 — Using `continue`

Write a function `sum_positive_even(nums)` that returns the sum of all numbers in a list that are both:
- positive
- even

Use a `while` loop and `continue`.

```python
def sum_positive_even(nums):
    """Return the sum of positive even numbers in nums."""
    idx = 0
    total = 0
    
    while idx < len(nums):
        value = nums[idx]
        idx += 1
        
        if value <= 0:
            continue
        if value % 2 != 0:
            continue
        
        total += value
    
    return total


assert sum_positive_even([1, 2, 3, 4, -6, 8]) == 14
assert sum_positive_even([-2, -4, 1, 3]) == 0
sum_positive_even([12, 7, 0, 6, -8, 10])
```

## Problem 7 — Search with `while ... else`

Write a function `find_or_append(values, target)` that:
- searches for `target` in the list using a `while` loop
- if found, returns its index
- if not found, appends it and returns the new index

Use `while ... else`.

```python
def find_or_append(values, target):
    """Return index of target if found; otherwise append and return new index."""
    idx = 0
    while idx < len(values):
        if values[idx] == target:
            return idx
        idx += 1
    else:
        values.append(target)
        return len(values) - 1


nums = [4, 7, 9]
assert find_or_append(nums, 7) == 1
assert nums == [4, 7, 9]
assert find_or_append(nums, 10) == 3
assert nums == [4, 7, 9, 10]
nums
```

## Problem 8 — Repeated division until condition fails

Write a function `halve_until_small(x, threshold)` that repeatedly divides `x` by `2` until `x <= threshold`.

Return the number of divisions performed.

Requirements:
- assume `x > 0` and `threshold > 0`
- if `x <= threshold` initially, return `0`

```python
def halve_until_small(x, threshold):
    """Return how many halvings are needed to make x <= threshold."""
    count = 0
    while x > threshold:
        x /= 2
        count += 1
    return count


assert halve_until_small(100, 12.5) == 3
assert halve_until_small(5, 10) == 0
halve_until_small(81, 5)
```

## Problem 9 — Integer square root by iteration

Write a function `integer_sqrt(n)` that returns the greatest integer `k` such that `k*k <= n`.

Examples:
- `integer_sqrt(16)` → `4`
- `integer_sqrt(20)` → `4`

Requirements:
- use a `while` loop
- do not use `math.sqrt`
- raise `ValueError` if `n < 0`

```python
def integer_sqrt(n):
    """Return the floor of the square root of a non-negative integer."""
    if n < 0:
        raise ValueError("n must be non-negative")
    
    k = 0
    while (k + 1) * (k + 1) <= n:
        k += 1
    return k


assert integer_sqrt(0) == 0
assert integer_sqrt(16) == 4
assert integer_sqrt(20) == 4
integer_sqrt(99)
```

## Problem 10 — Number guessing logic (simulated)

Write a function `guess_attempts(secret, guesses)` that simulates a guessing game.

Requirements:
- scan through `guesses` using a `while` loop
- return the number of attempts needed to guess `secret`
- if the secret is never guessed, return `-1`

Example:
- `guess_attempts(7, [2, 4, 7, 9])` should return `3`

```python
def guess_attempts(secret, guesses):
    """Return 1-based attempt count for the first correct guess, or -1 if absent."""
    idx = 0
    while idx < len(guesses):
        if guesses[idx] == secret:
            return idx + 1
        idx += 1
    return -1


assert guess_attempts(7, [2, 4, 7, 9]) == 3
assert guess_attempts(5, [1, 2, 3]) == -1
guess_attempts(42, [10, 20, 30, 42])
```

## Problem 11 — Nested `while` loops

Write a function `multiplication_table(n)` that returns an `n x n` multiplication table as a nested list.

Example for `n = 3`:

`[[1, 2, 3], [2, 4, 6], [3, 6, 9]]`

Use nested `while` loops.

```python
def multiplication_table(n):
    """Return an n x n multiplication table using nested while loops."""
    table = []
    row_num = 1
    
    while row_num <= n:
        row = []
        col_num = 1
        while col_num <= n:
            row.append(row_num * col_num)
            col_num += 1
        table.append(row)
        row_num += 1
    
    return table


assert multiplication_table(1) == [[1]]
assert multiplication_table(3) == [[1, 2, 3], [2, 4, 6], [3, 6, 9]]
multiplication_table(4)
```

## Problem 12 — Remove duplicates while preserving order

Write a function `deduplicate(values)` that returns a new list containing the first occurrence of each value, in order.

Requirements:
- use `while` loops
- do not use `set`
- preserve the original order

```python
def deduplicate(values):
    """Return a list with duplicates removed, preserving first occurrences."""
    result = []
    i = 0
    
    while i < len(values):
        value = values[i]
        j = 0
        seen = False
        
        while j < len(result):
            if result[j] == value:
                seen = True
                break
            j += 1
        
        if not seen:
            result.append(value)
        
        i += 1
    
    return result


assert deduplicate([1, 2, 1, 3, 2, 4, 3]) == [1, 2, 3, 4]
assert deduplicate([]) == []
deduplicate(["a", "b", "a", "c", "b"])
```

## Problem 13 — Prime check with `while ... else`

Write a function `is_prime(n)` that returns `True` if `n` is prime and `False` otherwise.

Requirements:
- use a `while` loop
- use `while ... else`
- `n < 2` should return `False`

```python
def is_prime(n):
    """Return True if n is prime, else False."""
    if n < 2:
        return False
    
    divisor = 2
    while divisor * divisor <= n:
        if n % divisor == 0:
            return False
        divisor += 1
    else:
        return True


assert is_prime(2) is True
assert is_prime(3) is True
assert is_prime(4) is False
assert is_prime(29) is True
is_prime(35)
```

## Problem 14 — Longest streak

Write a function `longest_non_decreasing_streak(nums)` that returns the length of the longest contiguous streak where each number is greater than or equal to the previous one.

Examples:
- `[1, 2, 2, 1, 3, 4, 0]` has longest non-decreasing streak length `3` (`1, 2, 2` or `1, 3, 4`)

Requirements:
- use a `while` loop
- return `0` for an empty list

```python
def longest_non_decreasing_streak(nums):
    """Return the longest contiguous non-decreasing streak length."""
    if not nums:
        return 0
    
    best = 1
    current = 1
    idx = 1
    
    while idx < len(nums):
        if nums[idx] >= nums[idx - 1]:
            current += 1
        else:
            if current > best:
                best = current
            current = 1
        idx += 1
    
    if current > best:
        best = current
    
    return best


assert longest_non_decreasing_streak([]) == 0
assert longest_non_decreasing_streak([5]) == 1
assert longest_non_decreasing_streak([1, 2, 2, 1, 3, 4, 0]) == 3
longest_non_decreasing_streak([4, 4, 4, 1, 2, 3, 3, 0])
```

## Problem 15 — Repeated subtraction division

Write a function `divide_by_subtraction(dividend, divisor)` that returns a tuple `(quotient, remainder)` using repeated subtraction.

Requirements:
- assume `dividend >= 0`
- assume `divisor > 0`
- do not use `//` or `%`
- use a `while` loop

```python
def divide_by_subtraction(dividend, divisor):
    """Return (quotient, remainder) via repeated subtraction."""
    quotient = 0
    remainder = dividend
    
    while remainder >= divisor:
        remainder -= divisor
        quotient += 1
    
    return quotient, remainder


assert divide_by_subtraction(17, 5) == (3, 2)
assert divide_by_subtraction(20, 4) == (5, 0)
divide_by_subtraction(43, 6)
```

## Problem 16 — Practical mini-project: ATM withdrawal simulation

Write a function `atm_withdrawals(balance, attempts)` that simulates withdrawal attempts from an account.

Each value in `attempts` is a requested withdrawal amount.

Rules:
- if an amount is less than or equal to 0, ignore it
- if an amount is greater than the current balance, stop processing immediately
- otherwise subtract it from the balance
- return the final balance and the number of successful withdrawals as a tuple `(balance, success_count)`

Use a `while` loop.

```python
def atm_withdrawals(balance, attempts):
    """Simulate withdrawals until funds are insufficient or attempts are exhausted."""
    idx = 0
    success_count = 0
    
    while idx < len(attempts):
        amount = attempts[idx]
        idx += 1
        
        if amount <= 0:
            continue
        if amount > balance:
            break
        
        balance -= amount
        success_count += 1
    
    return balance, success_count


assert atm_withdrawals(100, [20, 0, -5, 30, 60, 10]) == (50, 2)
assert atm_withdrawals(50, [10, 15, 20]) == (5, 3)
atm_withdrawals(80, [25, 25, 40, 5])
```
