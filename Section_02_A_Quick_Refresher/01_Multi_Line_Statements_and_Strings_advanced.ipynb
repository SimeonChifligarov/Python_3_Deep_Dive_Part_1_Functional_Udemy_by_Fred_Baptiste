## Best-practice reminders

- Prefer **implicit continuation** inside `()`, `[]`, and `{}` over backslashes.
- Use `\\` only when there is no cleaner alternative.
- Use triple-quoted strings for **real strings** such as templates, messages, and docstrings.
- Do **not** use triple-quoted strings as fake multi-line comments.
- Be careful: indentation inside triple-quoted strings is preserved.
- For teaching and debugging, use `repr(...)` when you need to see hidden newline and space characters clearly.


## Problem 1 — Predict the Result of Implicit Continuation

Without running the code, predict the final values of `numbers`, `config`, and `total`.

Also explain **why** the code is valid even though each statement spans multiple physical lines.

````python
numbers = [
    10,
    20,
    30,
    40,
]

config = {
    "host": "localhost",
    "port": 5432,
    "debug": True,
}

total = (
    numbers[0]
    + numbers[1]
    + numbers[2]
    + numbers[3]
)

print(numbers)
print(config)
print(total)
````

### Solution 1

````python
numbers = [
    10,
    20,
    30,
    40,
]

config = {
    "host": "localhost",
    "port": 5432,
    "debug": True,
}

total = (
    numbers[0]
    + numbers[1]
    + numbers[2]
    + numbers[3]
)

print(numbers)
print(config)
print(total)
````

**Explanation:**

This works because Python allows **implicit line continuation** inside brackets, braces, and parentheses. The interpreter waits until the enclosing `[]`, `{}`, or `()` is closed before ending the logical statement.

This is the preferred style because it is readable and less error-prone than explicit backslashes.

## Problem 2 — Repair a Broken Collection with Comments

The following code is intended to create a list, but it is broken.

Fix it using best practices so that it evaluates to `[1, 2, 3]`.

````python
values = [1,   # first value
          2,   # second value
          3    # third value ]
]  # TODO remove this
````

### Solution 2

````python
values = [
    1,  # first value
    2,  # second value
    3,  # third value
]

print(values)
````

**Explanation:**

The original code fails because the closing `]` appears inside the comment text. Once `#` starts a comment, the rest of that physical line is ignored.

Best practice: when using end-of-line comments inside a multi-line collection, place the closing bracket on its own line.

## Problem 4 — Replace Explicit Continuation with a Better Alternative

The code below works, but it does not follow best practice.

Rewrite it so that it avoids backslashes.

````python
is_valid = "python" in "I am learning python" and \
           len("notebook") > 3 and \
           10 < 20

print(is_valid)
````

### Solution 4

````python
is_valid = (
    "python" in "I am learning python"
    and len("notebook") > 3
    and 10 < 20
)

print(is_valid)
````

**Explanation:**

Wrapping the expression in parentheses is cleaner than using `\\`. Backslashes are fragile because stray spaces or comments can easily break them.

## Problem 5 — Understand Why a Backslash Version Fails

This code is invalid. Explain why, then fix it.

````python
result = 100 + \  # add base amount
         20

print(result)
````

### Solution 5

````python
# Best fix: avoid the backslash completely.
result = (
    100  # add base amount
    + 20
)

print(result)
````

**Explanation:**

A backslash used for explicit line continuation must be the last meaningful character on the line. Putting a comment after it breaks the continuation.

Again, the best repair is usually to replace explicit continuation with implicit continuation.

## Problem 7 — Remove Unwanted Indentation from a Multi-Line String

The function below returns a multi-line message, but the lines after the first one contain unwanted leading spaces.

Rewrite the function so the returned string is:

```text
Dear user,
Your report is ready.
Please review it.
```

without unwanted indentation.

````python
def build_message():
    return """Dear user,
    Your report is ready.
    Please review it."""

print(build_message())
print(repr(build_message()))
````

### Solution 7

````python
def build_message():
    return (
        "Dear user,\n"
        "Your report is ready.\n"
        "Please review it."
    )

print(build_message())
print(repr(build_message()))
````

**Why this solution is strong:**

For short fixed text where you want exact control, adjacent string literals inside parentheses are often clearer than an indented triple-quoted string.

They avoid accidental indentation while still being readable.

## Problem 8 — Build a SQL Query Readably

Construct the following exact string in a readable multi-line way:

```sql

FROM students
WHERE score >= 90
ORDER BY score DESC
```

Then print both the string and its `repr(...)`.

### Solution 8

## Problem 9 — Distinguish Real Comments from Triple-Quoted Strings

What is the practical problem with using a triple-quoted string as a fake comment inside a function?

Refactor the function to use proper comments.

````python
def compute_area(width, height):
    """Multiply width by height after basic validation."""
    """
    This block is intended as a comment,
    but it is actually a string literal.
    """
    if width < 0 or height < 0:
        raise ValueError("Dimensions must be non-negative")
    return width * height

print(compute_area(4, 5))
````

### Solution 9

````python
def compute_area(width, height):
    """Multiply width by height after basic validation."""
    # Validate that dimensions are not negative.
    # These are real comments, not unused string literals.
    if width < 0 or height < 0:
        raise ValueError("Dimensions must be non-negative")
    return width * height

print(compute_area(4, 5))
````

**Explanation:**

A stray triple-quoted block is still a real string literal in the compiled code. It is not the same as a comment. Only the first such string in a module, class, or function body has a special role as a docstring.

For comments, use `#`.

## Problem 12 — Create a Docstring and Inspect It

Write a function `describe()` whose docstring is a proper multi-line string explaining what the function does.

Then print `describe.__doc__`.

### Solution 12

````python
def describe():
    """Return a short status message.

    This function demonstrates a real use of a multi-line string:
    a function docstring.
    """
    return "System ready"

print(describe.__doc__)
````

**Key idea:**

Triple-quoted strings are appropriate here because a docstring is an actual runtime string attached to the function.

## Challenge Problem 13 — Debug Hidden Characters

A programmer expects the following string to be exactly `'A\nB\nC'`, but it is not.

Use `repr(...)` to diagnose the problem, then fix it.

````python
letters = """A
    B
    C"""

print(repr(letters))
````

### Solution 13

````python
letters = """A
    B
    C"""
print("Broken version:", repr(letters))

letters = "A\nB\nC"
print("Fixed version: ", repr(letters))
````

**Explanation:**

The triple-quoted version preserved the indentation before `B` and `C`. The corrected version uses explicit newline escapes to guarantee the exact string content.
